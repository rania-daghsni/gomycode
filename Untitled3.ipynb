```python
import random
x=int(input("write your guess"))
y=random.randint(1,100)
print(y)
while x!=y:
  if(x>y):
    print("your guess is very high")
  else:
    print("your guess is very low")
  x=int(input("write your guess"))
print("correct!")
```
