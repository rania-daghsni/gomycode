```python
def calcul(n,m):
  n=int(input('Enter the first number:'  ))
  m=int(input('Enter the second number :'))
  o=input('Enter the operator (+, -, *, /)')
  if o=='+':
    return(n+m)
  elif o=='-':
    return(n-m)
  elif o=='*':
    return(n*m)
  elif o=='/'  and m!=0:
    return(n/m)
calcul(n,m)

```
