```python
class  Point3D():
  def __init__(self,x,y,z):
    self.x=x
    self.y=y
    self.z=z
  def format(self):
    return(self.x, self.y ,self.z)
my_point=Point3D(1,2,3)
print(my_point.format())
```

```python
class  Rectangle():
  def __init__(self,length,width):
    self.length=length
    self.width=width
  def area(self):
    return (self.length*self.width)
  def perimeter(self):
    return (self.length+self.width)*2
my_rectangle=Rectangle(4,3)
print(my_rectangle.area())
print(my_rectangle.perimeter())
```

```python
class bank():
  def __init__(balance,deposit,withdraw):
    self.balance=balance
    self.deposit=deposit
    self.withdraw=withdraw
  def deposit (self):
    return(self.balance+self.deposit)
  def withdraw(self):
    return(self.balance-self.withdraw)
new_balance=bank(1200,100,20)
print(new_balance.deposit())
print(new_balance.withdraw())

```
